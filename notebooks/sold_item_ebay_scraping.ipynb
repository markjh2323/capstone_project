```python
import requests
import bs4 as BeautifulSoup
import pandas as pd
import re
```

```python
def get_page(url):
    response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'}
)
    
    if not response.ok:
        print('Server responded:', response.status_code)
    else:
        soup = BeautifulSoup.BeautifulSoup(response.text, 'lxml')
    return soup
```

```python



def get_detail_data(soup):
    """
    Scraping info from listing: sold title, price, number sold, shipping_type, condition, condition description.
    """
    try:
        sold_title = soup.body.find('div', id='mainContent').find('h1', class_='it-ttl').text[16:]
    except:
        sold_title = None

    try:
        try:
            sold_price = float(re.findall(r"[-+]?\d*\.\d+|\d+", soup.body.find('span', id='prcIsum').text)[0])
        except:
            sold_price = float(re.findall(r"[-+]?\d*\.\d+|\d+", soup.body.find('span', id='mm-saleDscPrc').text)[0])
    except:
        sold_price = None

    try:
        sold_shipping_type = float(re.findall(r"[-+]?\d*\.\d+|\d+", test_soup.body.find('span', id='fshippingCost').text.replace('\nFREE\n', '0'))[0])
    except:
        sold_shipping_type = None

    try:
        sold_condition = soup.body.find('div', id='vi-itm-cond').text
    except:
        sold_condition = None

    try:
        condition_description = soup.body.find('div', class_='ux-layout-section__item ux-layout-section__item-background ux-layout-section__item--table-view').find_all('div', 'ux-layout-section__row')[1].find('div', class_='ux-labels-values__values').span.text.lower()
    except:
        condition_description = None
     
    try:
        sold_num_reviews = int(re.findall(r"[-+]?\d*\.\d+|\d+",soup.body.find('span', class_='mbg-l').text)[0])

    except:
        sold_num_reviews = None
        
    data = {
        'sold_title': sold_title,
        'sold_price': sold_price,
        'sold_shipping_type': sold_shipping_type,
        'sold_condition': sold_condition,
        'sold_num_reviews': sold_num_reviews
    }
    return data
    
def get_index_data(soup):
    try:
        links = soup.find_all('a', class_='s-item__link')
    except:
        links = []
    urls = [item.get('href') for item in links]
    
    return urls



# Sold items functions to extract titles, prices, conditions, etc.
def get_titles(soup):
    try:
        raw_titles = soup.find_all('h3', class_='s-item__title s-item__title--has-tags')
    except:
        raw_titles = []
    titles_list = [item.text for item in raw_titles]
    
    return titles_list

def get_conditions(soup):
    try:
        raw_conditions = soup.find_all('span', class_='SECONDARY_INFO')
    except:
        raw_titles = []
    conditions_list = [item.text for item in raw_conditions]
    
    return conditions_list

def get_prices(soup):
    try:
        raw_prices = soup.find_all('span', class_='s-item__price')
    except:
        raw_prices = []
    prices_list = [item.text for item in raw_prices]
    
    return prices_list

def get_dates_sold(soup):
    try:
        raw_dates = soup.find_all('div', class_='s-item__title--tagblock')
    except:
        raw_dates = []
    dates_list = [item.text for item in raw_dates]

    return dates_list




```
